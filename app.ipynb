```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np
from jaal import Jaal
```

```python
path="data/twitter/"
```

```python
circle_78813_file = open(f'{path}100318079.circles')
circle_78813 = circle_78813_file.readlines()
circle_78813 = [c.strip("\n").split("\t") for c in circle_78813]
```

```python
edge_78813_file = open(f'{path}100318079.edges')
edge_78813 = edge_78813_file.readlines()
edge_78813 = [c.strip("\n").split(" ") for c in edge_78813]
```

```python
# checking weather it includes duplicates
unique = []
dup=[]
for arr in circle_78813:
    for i in range(1, len(arr)):
        if arr[i] in unique:
            dup.append(arr[i])
        else:
            unique.append(arr[i])
print(dup)
```

```python
following_7813 =[]
for arr in circle_78813:
    for i in range(1, len(arr)- 1):
        if i == len(arr) - 1:
            following_7813.append([arr[i], arr[1]])
        else:
            following_7813.append([arr[i], arr[i+1]])
following_7813.extend(edge_78813 )
following_7813_df = pd.DataFrame(columns = ["from", "to"], data = following_7813)
following_7813_df = following_7813_df.drop_duplicates()
following_7813_df.head()
```
